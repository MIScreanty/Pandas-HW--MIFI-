#### Объединение данных
Так как train.csv содержит основную информацию об интеракциях пользователей, а questions.csv и lectures.csv — метаданные о вопросах и лекциях, мы можем объединить их по полю content_id, чтобы добавить соответствующую информацию для дальнейшего анализа.

```python
# Объединеним вопросы по content_id (только для строк, где content_type_id = 0)
train_questions_df = train_df[train_df['content_type_id'] == 0].merge(questions_df, left_on='content_id', right_on='question_id', how='left')
```

```python
train_questions_df.head(5)
```

```python
train_questions_df.describe(include='all')
```

```python
# Объединим лекции по content_id (только для строк, где content_type_id = 1)
train_lectures_df = train_df[train_df['content_type_id'] == 1].merge(lectures_df, left_on='content_id', right_on='lecture_id', how='left')
```

```python
train_lectures_df.head(5)
```

#### Проведем анализ и визуализацию результата по характеристикам

```python

# Импортируем библиотеку для визуализации из библиотеки Matplotlib
import matplotlib.pyplot as plt
```

##### 1. Изучим как различные разделы (part) влияют на правильность ответов.

```python
part_performance = train_questions_df.groupby('part')['answered_correctly'].mean()
```

```python
part_performance.plot(kind='bar', color='skyblue')
plt.title("Успеваемость по разделам теста")
plt.xlabel("Раздел")
plt.ylabel("Средняя правильность ответов")
plt.show()
```

Вывод: раздел 1 имеет самую высокую успеваемость, а раздел 5 - самую низкую.

##### 2. Исследуем влияет ли наличие объяснений после вопроса на результаты

```python
train_df['prior_question_had_explanation'] = train_df['prior_question_had_explanation'].fillna(False)
explanation_performance = train_df.groupby('prior_question_had_explanation')['answered_correctly'].mean()
```

```python
explanation_performance.plot(kind='bar', color=['salmon', 'lightgreen'])
plt.title("Успеваемость в зависимости от наличия объяснений")
plt.xlabel("Были объяснения?")
plt.ylabel("Средняя правильность ответов")
plt.xticks([0, 1], ['Нет', 'Да'], rotation=0)
plt.show()
```

Вывод: средняя правильность ответа при наличии объяснений значительно выше чем при их отсутствии.

##### 3. Проанализируем как среднее время, затраченное на ответ, влияет на его правильность.

```python
# Преобразуем время ответа из миллисекунд в минуты
train_df['prior_question_elapsed_time_minutes'] = train_df['prior_question_elapsed_time'] / 60000
```

```python
# Определяем 10 равномерных интервалов от минимального до максимального значения времени в минутах
bins = np.linspace(train_df['prior_question_elapsed_time_minutes'].min(),
                   train_df['prior_question_elapsed_time_minutes'].max(), 11)

# Группируем по этим интервалам и рассчитываем среднюю правильность ответов
train_df['time_bins'] = pd.cut(train_df['prior_question_elapsed_time_minutes'], bins=bins)
elapsed_time_performance = train_df.groupby('time_bins')['answered_correctly'].mean()

# Для удобства создаем метки на графике
elapsed_time_performance.index = [f"{int(bin.left)} - {int(bin.right)} мин" for bin in elapsed_time_performance.index]
```

```python
elapsed_time_performance.plot(kind='line', marker='o', color='purple')
plt.title("Успеваемость в зависимости от времени ответа (интервалы от мин до макс)")
plt.xlabel("Время ответа (минуты)")
plt.ylabel("Средняя правильность ответов")
plt.xticks(rotation=45)
plt.show()
```

Вывод: Самая высокая средняя правильность ответа наблюдается при скорости ответа до 2х минут.
